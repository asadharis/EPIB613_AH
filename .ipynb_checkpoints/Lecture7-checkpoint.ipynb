### 7.2.2 Numerical representation
#### Histogram

```r
par(mfrow=c(2 ,2))
hist(tg$len,
     main = "Distribution of Tooth Length",
     xlab = "Length",
     freq = F,   # F gives probability, T gives count
     breaks = 2)
hist(tg$len,
     main = "Distribution of Tooth Length",
     xlab = "Length",
     freq = F,
     breaks = 5)
hist(tg$len,
     main = "Distribution of Tooth Length",
     xlab = "Length",
     freq = F,
     breaks = 20)
hist(tg$len,
     main = "Distribution of Tooth Length",
     xlab = "Length",
     freq = F,
     breaks = 100)
rug(tg$len)   # Add ticks on the x axis.

# See how different number of bins can affect our arbitrary judgement of the distribution.
```
